```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import cv2  #arquivo de imagem
import random
import zipfile #faz descompactação
```

```python
# all images are load in variable x...
Arq =['cardboard.zip','glass.zip','metal.zip','not_recycle.zip',
      'paper.zip','plastic.zip']

img_name=[]
x_gray=[]
x_original = []
for FILE in Arq:  #essa é a diferença, carrega todos os arquivos 
  file_name = zipfile.ZipFile(FILE, 'r')
  file_name.extractall()
  with zipfile.ZipFile(FILE, "r") as f:
      for name in f.namelist():
          img_name.append(name)
          #x_gray.append(cv2.imread(name, 0))
          x_gray.append(cv2.resize(cv2.imread(name, 0),(80,60)))   #carrega em escala cinza e resolução (80,60)
```

```python
y_names = []
for yy in img_name:
  name = yy[0]+yy[1]+yy[2]
  y_names.append(name)   #colocando os nomes dos arquivos apenas com as 3 primeiras letras
```

```python
unique_list = []    
for yy in y_names:
  if yy not in unique_list:
    unique_list.append(yy)
```

```python
n = len(unique_list)
y = []
for yy in y_names:
  for i in range(n):
    if(yy == unique_list[i]):
      y.append(i)
```

```python
# view some images from the train_ds
num = len(y_names)    #lista com a quantidade de fotos
plt.figure(figsize=(16, 16))
for i in range(12):
  ax = plt.subplot(4, 4, i + 1)
  kk = random.randrange(num) 
  plt.imshow(x_gray[kk], cmap='gray')
  plt.title(y_names[kk])
  plt.axis("off")
```

```python
x_data = []
for xx in x_gray:
  x_data.append(xx.ravel())   #converte a matriz em uma linha única na lista em branco criada

df = pd.DataFrame(x_data) #converte a lista em dataframe
print(df)
```
